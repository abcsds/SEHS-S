# Analisis exploratorio de datos

Leer base de datos y guardarla en variable db (DataBase). Las variables en R se llaman Data Frames, o marcos de datos, que representan matrices de valores y sus metadotos.

```r
db <- read.csv("Depurada.csv") # Se lee el archivo y se guarda en la variable db
head(db) # Mostrar los primeros datos de la variable
```

```r
nrow(db) # Cuantas observaciones tiene la base de datos? Number of rows
```

Crear Dataframe db12 y copiar columnas de db a db12

```r
db12 <- data.frame(row.names = 1:nrow(db)) # Crear un nuevo dataframe de 2033 filas
db12$Grado <- db$Grado # Copiar las variables que deben de permanecer iguales
db12$Sexo <- db$Sexo
db12$Edad <- db$Edad
```

Sacar promedio de las 36 variables de db y crear una columna en db12.
Se deben remover los resultados que no tengan valores `na.rm = TRUE` para que no se propaguen los `NA`

```r
db12$Autoconsciencia<-rowMeans(subset(db, select = c(Autoconsciencia1,Autoconsciencia2,Autoconsciencia3)), na.rm = T)
db12$Eficacia<-rowMeans(subset(db, select = c(Eficacia1,Eficacia2,Eficacia3)), na.rm = T)
db12$Persistencia<-rowMeans(subset(db, select = c(Persistencia1,Persistencia2,Persistencia3)), na.rm = T)
db12$ApoyoEscuela<-rowMeans(subset(db, select = c(ApoyoEscuela1,ApoyoEscuela2,ApoyoEscuela3)), na.rm = T)
db12$CoherenciaFam<-rowMeans(subset(db, select = c(CoherenciaFam1,CoherenciaFam2,CoherenciaFam3)), na.rm = T)
db12$Compañerismo<-rowMeans(subset(db, select = c(Compañerismo1,Compañerismo2,Compañerismo3)), na.rm = T)
db12$RegulaciónEmo<-rowMeans(subset(db, select = c(RegulaciónEmo1,RegulaciónEmo2,RegulaciónEmo3)), na.rm = T)
db12$Empatía<-rowMeans(subset(db, select = c(Empatía1,Empatía2,Empatía3)), na.rm = T)
db12$ControlConducta<-rowMeans(subset(db, select = c(ControlConducta1,ControlConducta2,ControlConducta3)), na.rm = T)
db12$Optimismo<-rowMeans(subset(db, select = c(Optimismo1,Optimismo2,Optimismo3)), na.rm = T)
db12$Entusiasmo<-rowMeans(subset(db, select = c(Entusiasmo1,Entusiasmo2,Entusiasmo3)), na.rm = T)
db12$Gratitud<-rowMeans(subset(db, select = c(Gratitud1,Gratitud2,Gratitud3)), na.rm = T)
```

La nueva base de datos queda con la siguiente forma:

```r
head(db12) # Mostrar los primeros datos
variables <- names(db12[,4:15]) # Guardar los nombres de las columnas en una variable
```

```r
summary(db12) # Esta función da información sobre la dispersión de cada variable
```

Se encuentra la media de cada una de las columnas

```r
colMeans(db12) # Colmeans = Media de cada columna
```

Y la desviacicón estandar de cada columna

```r
apply(db12,2,sd) # Aplicar la función sd() a cada columna de db12
```

```r
cbind(colMeans(db12), apply(db12,2,sd)) # Una tabla con promedios y desviaciones estandar
```

Las gráficas boxplots perimten ver disperción de los datos, así como sus cuartiles y outliers. En este caso cada una de las variables presenta una distribución característica entre 0 y 5. 

```r
bp <- boxplot(db12[,4:15], names = F, cex.axis=1, xaxt='n')
text(seq(1, 12, by=0.9), par("usr")[3] - 0.48, labels = names(db12[,4:15]), srt = 90, pos = 1, adj = 1, xpd = TRUE)
#par(mar = c(6,4,3,2))
```
